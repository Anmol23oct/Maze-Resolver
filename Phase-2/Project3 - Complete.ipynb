```python
from queue import PriorityQueue
import operator
import sys
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
#without Dictionary update
        

class AStarResolver:
    def __init__(self,Klst,n,m,dlst):
        #self.path=[]
        self.priorityQueue=PriorityQueue()
        self.lst=Klst
        self.n=n
        self.m=m
        self.Klst=Klst
        self.cells_processed=0
        self.exploredset={()}
        
    
    def init_dict(self,x = 0,y = 0):
        self.d = {(x,y): [ 0, None, None]}
        return self.d

    def Solve(self,sx = 0,sy = 0,tx=0,ty=0):
        found=0
        self.priorityQueue.put((0,sx,sy))
        self.init_dict(sx,sy)
        while not self.priorityQueue.empty():
            
            
            #extracting start 2d list from tuple
            
            
            temp=self.priorityQueue.get()
            #print("PQ top=",temp)
            nTemp=temp[1]
            mTemp=temp[2]
            self.cells_processed=self.cells_processed+1
            #print(self.priorityQueue.empty())
            if nTemp==tx and mTemp==ty:
                
                found=1
                #print("found")
                break
                
                
            if((nTemp, mTemp) not in self.d):
                return 
            g=self.d[nTemp,mTemp][0]
            g=g+1

            
            if nTemp-1>=0:
                up=(nTemp-1,mTemp)
                if self.lst[int(nTemp)-1][int(mTemp)]!=-1:
                    f1=self.calculateDist(up,g)
                    if (nTemp-1,mTemp) not in self.d:
                        self.d[nTemp-1,mTemp]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f1[0],f1[1],f1[2]))
                        #self.exploredset.update([(nTemp-1,mTemp)])
                    elif self.d[nTemp-1,mTemp][0]>g:
                        self.d[nTemp-1,mTemp]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f1[0],f1[1],f1[2]))
                
                        
                
                    
                    
            
            if nTemp+1<self.n:
                down=(nTemp+1,mTemp)
                if self.lst[int(nTemp)+1][int(mTemp)]!=-1:
                    f2=self.calculateDist(down,g)
                    if (nTemp+1,mTemp) not in self.d:
                        self.d[nTemp+1,mTemp]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f2[0],f2[1],f2[2]))
        
                    elif self.d[nTemp+1,mTemp][0]>g:
                        self.d[nTemp+1,mTemp]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f2[0],f2[1],f2[2]))
            
                        
                
            
            if mTemp+1<self.m:
                east=(nTemp,mTemp+1)
                if self.lst[nTemp][mTemp+1]!=-1:
                    f3=self.calculateDist(east,g)
                    if (nTemp,mTemp+1) not in self.d:
                        self.d[nTemp,mTemp+1]= [g,nTemp,mTemp]
                        self.priorityQueue.put((f3[0],f3[1],f3[2]))

                    elif self.d[nTemp,mTemp+1][0]>g:
                        self.d[nTemp,mTemp+1]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f3[0],f3[1],f3[2]))
    
                
            if mTemp-1>=0:
                west=(nTemp,mTemp-1)
                if self.lst[nTemp][mTemp-1]!=-1:
                    f4=self.calculateDist(west,g)
                    if (nTemp,mTemp-1) not in self.d:
                        self.d[nTemp,mTemp-1]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f4[0],f4[1],f4[2]))
                        
                    elif self.d[nTemp,mTemp-1][0]>g:
                        self.d[nTemp,mTemp-1]=[g,nTemp,mTemp]
                        self.priorityQueue.put((f4[0],f4[1],f4[2]))
            
            
            
                
            
        if found==0:
            ##print ("NOSUCCESS")
            self.EmptyQueue()
            filst=[]
            return filst
                
        else:
            result = self.PathIterator(tx,ty)
            self.EmptyQueue()
            return result
            
            
            
            
            
            
            
        
        
        
    def PathIterator(self,tx,ty):
        filst=[]
        filst.append((tx,ty))
        #print(self.d[self.n-1,self.m-1])
        Pn=self.d[tx,ty][1]
        Pm=self.d[tx,ty][2]
        while Pn is not None or Pm is not None:
            filst.append((Pn,Pm))
            #print("Path Total[{},{}]".format(Pn, Pm),end=" ")
            An=self.d[Pn,Pm][1]
            Am=self.d[Pn,Pm][2]
            Pn,Pm=An,Am
        filst.reverse()
        #print(filst)
        return filst
    
    def EmptyQueue(self):
        while not self.priorityQueue.empty():
            try:
                self.priorityQueue.get(False)
            except Empty:
                continue
            self.priorityQueue.task_done()        
        
            
            
        
        
       
    def calculateDist(self,cord,g):
        
        h=abs(self.n-cord[0])+abs(self.m-cord[1])  #Manhattan distance
        
        #h=math.sqrt(pow(self.n-cord[0],2)+pow(self.m-cord[1],2)) #euclidean distance
        
        #h=max(abs(self.n-cord[0]),abs(self.m-cord[1])) # chebychef
        
        return (g+h,cord[0],cord[1])
    
    
    def get_cells_processed(self):
        return self.cells_processed
```

# Agent 6

```python
class Phases:
    
    def __init__(self,Klst,Flst,n,m,target):
        #self.path=[]
        self.exploredset={()}
        self.book={}
        self.Klst=Klst
        self.Flst=Flst
        self.n=n
        self.m=m
        self.trajlst = []
        self.dKprob={(0,0): [ 0, "NA"]}  #{(x,y): [prob,terraintype]} 1:Flat 2:Hilly 3:ThickForest
                                         #terrain type- Hilly - H, Flat - F, ThickForest- TF
        self.target=target
        self.fnrH=0.5 #FNR - hilly 
        self.fnrF=0.2 #FNR - Flat
        self.fnrTF=0.8 #FNR - Thick Forest
        self.init_dict()
    
    
    def init_dict(self,x = 0,y = 0):
        initProb=1/(self.n*self.n)
        for i in range(self.n):
            for j in range(self.m):
                self.dKprob[i,j]=[initProb,"NA"]
        
        #Initializing Dictionary with default values, Here "NA" is terrain not defined.
    
    #Planning Phase
    def Planning(self,source,target):
        
        # calling A star with source and max probability target
        a=AStarResolver(self.Klst,self.n,self.m,self.Klst)
        path=a.Solve(source[0],source[1],target[0],target[1])
        
        # If we didn't get any path from a star to a particular assumed target then we will take target probability as 0 
        if len(path)==0:
            self.updateKnowledgeBase(target[0],target[1],"NA","B")
        return path
    
    #Execution Phase
    def Execution(self,path):
        compLoop=True
        bcell=(0,0)
        ubcell=(0,0)
        movement=0
        
        for item in range(len(path)):
            
            # if we get the blocked cell while iterating planned path
            if len(path)>1 and item!=0:
                x=path[item][0]
                y=path[item][1]
                if self.Flst[x][y]==-1:
                    self.Klst[x][y]=-1
                    # update whole grid when the cell is blocked
                    self.updateKnowledgeBase(x,y,"NA","B")
                    compLoop=False
                    bcell=path[item-1]
                    movement=movement+1
                    break

                else:

                    #Agent 6
                    # if we get the unblocked cell while iterating the path
                    #updating terrains without affecting probability
                    ubcell=(x,y)
                    if (x,y) in self.dKprob:
                        initProb=self.dKprob[x,y]
                        initProb=self.dKprob[x,y][0]
                        movement=movement+1 #Total Moves Counter
                        self.dKprob[x,y]=[initProb,self.IdentifyTerrain(self.Flst[x][y])]

               
        x=ubcell[0]
        y=ubcell[1]
        #Examination of the cell starts
        if compLoop==False:# cell was blocked
            return bcell,True,movement
        elif ubcell==self.target:#cell was unblocked and find the target
            
            #Random Generator function was called in order to compare its value with FNR of terrain
            val=self.generateFalseTarget()
            targetFound=self.TargetMeasure(val,self.IdentifyTerrain(self.Flst[x][y]))
            # If random value greater than FNR value of terrain for that particular cell then target found is true
            if targetFound==True:
                return (0,0),False,movement
            else:
                #If random value is lesser than FNR value of that particular cell then target not found
                #whole grid will be updated
                self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
                return ubcell,True,movement
        else:
            # cell was unblocked and target not found
            self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
            return ubcell,True,movement#last element
    
    #1st argument- x coordinate
    #2nd argument- y coordinate
    #3rd argument- terrain
    #4th Condition- Blocked,Unblocked
    def updateKnowledgeBase(self,x,y,terrain,cond):
        
        initProb=self.dKprob[x,y][0]
        #print("initProbbefore=",initProb)
        fnrH=0.5
        fnrF=0.2
        fnrTF=0.8
        if cond=="B":
            sumt=0
            #Blocked cell probability would be 0 and all other grid probabilities will change accordingly
            self.dKprob[x,y]=[0,terrain]
            for i in range(self.n):
                for j in range(self.m):
                    sumt=sumt+self.dKprob[i,j][0]
            
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]
                        self.dKprob[i,j]=[probij/sumt,terraini]
            
        elif cond=="UB":
            #Unbloced cell with target not found, will update probabilities of whole grid including unblocked cell
            if terrain =="F":
                probxy=initProb*fnrF
            elif terrain =="H":
                probxy=initProb*fnrH
            elif terrain=="TF":
                probxy=initProb*fnrTF
            sumt=0
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        sumt=sumt+self.dKprob[i,j][0]

            probxy=probxy/(sumt+probxy)
            self.dKprob[x,y]=[probxy,terrain]
            sumt=sumt+probxy
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]  
                        self.dKprob[i,j]=[probij/sumt,terraini]
                    
        
        
    # Return target with max probability and closest distance 
    def TargetwithMaxProb(self):
        maxed=max(self.dKprob, key=self.dKprob.get)
        return maxed
        
    # Identifying Terrain Types
    def IdentifyTerrain(self, terrain):
        if terrain==1:
            return "F"
        elif terrain==2:
            return "H"
        elif terrain==3:
            return "TF"
        else:
            return "F"
    # Generating random value between 0 and 1
    def generateFalseTarget(self):
        return random.uniform(0,1)
    
    # Calculating if random generated value is greater than FNR of terrain in order to find the target
    def TargetMeasure(self,val,terrain):
        if terrain=="F":
            if val>self.fnrF:
                return True
            else:
                return False
        elif terrain=="H":
            if val>self.fnrH:
                return True
            else:
                return False
        else:
            if val>self.fnrTF:
                return True
            else:
                return False

        
```

```python
def CreateRandom(n=3):
    import sys
    import numpy as np
    import random
    import pandas as pd
    a=[1,2,3,-1]
    mat=np.random.choice(a,size=(n,n),p = [0.3,0.3,0.3,0.1])
    mat[0,0] = 1
    targetx=int(random.uniform(0,n-1))
    targety=int(random.uniform(0,n-1))
    target=(targetx,targety)
    mat[targetx][targety]=1
    return mat,target
```

# Agent7

```python
class Phases7:
    
    def __init__(self,Klst,Flst,n,m,target):
        #self.path=[]
        self.exploredset={()}
        self.book={}
        self.Klst=Klst
        self.Flst=Flst
        self.n=n
        self.m=m
        self.trajlst = []
        self.dKprob={(0,0): [ 0, "NA"]}  #{(x,y): [prob,terraintype]} 1:Flat 2:Hilly 3:ThickForest
        #terrain type- Hilly-H, Flat - F, ThickForest- TF
        self.target=target
        self.fnrH=0.5 #FNR - hilly 
        self.fnrF=0.2 #FNR - Flat
        self.fnrTF=0.8 #FNR - Thick Forest
        self.init_dict()
    
    
    def init_dict(self,x = 0,y = 0):
        #initial probability of finding success
        unvisitedSucceeds= (1-self.fnrH)+(1-self.fnrF)+(1-self.fnrTF)
        initProb=(1/(self.n*self.n))*unvisitedSucceeds 
        for i in range(self.n):
            for j in range(self.m):
                self.dKprob[i,j]=[initProb,"NA"]
         
        
    def Planning(self,source,target):
        
        a=AStarResolver(self.Klst,self.n,self.m,self.Klst)
        path=a.Solve(source[0],source[1],target[0],target[1])
        if len(path)==0:
            self.updateKnowledgeBase(target[0],target[1],"NA","B")
        return path
    
    def Execution(self,path):
        compLoop=True
        bcell=(0,0)
        ubcell=(0,0)
        movement=0
        MaxprobChangedFlag=False
        for item in range(len(path)):
            if len(path)>1 and item!=0:
                x=path[item][0]
                y=path[item][1]
                if self.Flst[x][y]==-1:
                    self.Klst[x][y]=-1
                    self.updateKnowledgeBase(x,y,"NA","B")
                    compLoop=False
                    bcell=path[item-1]
                    movement=movement+1
                    break

                else:

                    ubcell=(x,y)
                    if (x,y) in self.dKprob:
                        initProb=self.dKprob[x,y]
                        initProb=self.dKprob[x,y][0]
                        terrain=self.dKprob[x,y][1]
                        if terrain =="NA":#First time you are discovering the terrain
                            
                            self.dKprob[x,y]=[initProb,self.IdentifyTerrain(self.Flst[x][y])]
                            if item!=(len(path)-1):
                                # We are changing the probability of successfully finding the target 
                                # when we observe the terrain
                                self.SucceedingProbability(x,y)
                            
                        else:
                            self.dKprob[x,y]=[initProb,self.IdentifyTerrain(self.Flst[x][y])]
                    movement=movement+1
                    
            # we are checking if our target still having maximum probability, if it doesn,t then replaning occurs
                    maxed=self.TargetwithMaxProb()
                    if maxed!=path[len(path)-1]:
                        MaxprobChangedFlag=True
                        compLoop=False
                        break
                        

        x=ubcell[0]
        y=ubcell[1]
        if compLoop==False:#cell was blocked or we get the cell with max probability other than our assumed target
            if MaxprobChangedFlag==True:
                
                return ubcell,True,movement
            else:
                
                return bcell,True,movement
            
        elif ubcell==self.target:#cell was unblocked and find the target
            val=self.generateFalseTarget()
            targetFound=self.TargetMeasure(val,self.IdentifyTerrain(self.Flst[x][y]))
            if targetFound==True:
                return (0,0),False,movement
            else:
                self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
                #Calculating probability of successfully finding the target
                self.SucceedingProbability(x,y)
                return ubcell,True,movement
    
        else:# cell was unblocked and target not found
            self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
            #Calculating probability of successfully finding the target
            self.SucceedingProbability(x,y)
            return ubcell,True,movement#last element
    
    #1st argument- x coordinate
    #2nd argument- y coordinate
    #3rd argument- terrain
    #4th Condition- Blocked,Unblocked
    def updateKnowledgeBase(self,x,y,terrain,cond):
        
        initProb=self.dKprob[x,y][0]
        fnrH=0.5
        fnrF=0.2
        fnrTF=0.8
        if cond=="B":
            sumt=0
            self.dKprob[x,y]=[0,terrain]
            for i in range(self.n):
                for j in range(self.m):
                    sumt=sumt+self.dKprob[i,j][0]
            
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]
                        self.dKprob[i,j]=[probij/sumt,terraini]
            
        elif cond=="UB":
            if terrain =="F":
                probxy=initProb*fnrF
            elif terrain =="H":
                probxy=initProb*fnrH
            elif terrain=="TF":
                probxy=initProb*fnrTF
            sumt=0
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        sumt=sumt+self.dKprob[i,j][0]
            #sumt=sumt+probxy
            probxy=probxy/(sumt+probxy)
             #xy prob update
            self.dKprob[x,y]=[probxy,terrain]
            sumt=sumt+probxy
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]  
                        self.dKprob[i,j]=[probij/sumt,terraini]
                    
    
    # We use this function in order to find the probability of successfully finding the target and update dictionary
    def SucceedingProbability(self,x,y):
        initprob=self.dKprob[x,y][0]
        terrain=self.dKprob[x,y][1]
        newProb=initprob
        if terrain=="H":
            newProb=initprob*(1-self.fnrH)
        elif terrain=="F":
            newProb=initprob*(1-self.fnrF)
        else:
            newProb=initprob*(1-self.fnrTF)
        
        self.dKprob[x,y]=[newProb,terrain]
            
    
    def TargetwithMaxProb(self):
        maxed=max(self.dKprob, key=self.dKprob.get)
        return maxed
        
    def IdentifyTerrain(self, terrain):
        if terrain==1:
            return "F"
        elif terrain==2:
            return "H"
        elif terrain==3:
            return "TF"
        else:
            return "F"
   
    def generateFalseTarget(self):
        return random.uniform(0,1)
    
    def TargetMeasure(self,val,terrain):
        if terrain=="F":
            if val>self.fnrF:
                return True
            else:
                return False
        elif terrain=="H":
            if val>self.fnrH:
                return True
            else:
                return False
        else:
            #print("in TF loop")
            if val>self.fnrTF:
                return True
            else:
                return False

        
```

```python
class Phases8A:
    
    def __init__(self,Klst,Flst,n,m,target):
        #self.path=[]
        self.exploredset={()}
        self.book={}
        self.Klst=Klst
        self.Flst=Flst
        self.n=n
        self.m=m
        self.trajlst = []
        self.dKprob={(0,0): [ 0, "NA"]}  #{(x,y): [prob,terraintype]} 1:Flat 2:Hilly 3:ThickForest
        #terrain type- Hilly-H, Flat - F, ThickForest- TF
        #self.decideDict={(0,0): [ 0, "NA"]}
        self.target=target
        self.fnrH=0.5 #FNR - hilly 
        self.fnrF=0.2 #FNR - Flat
        self.fnrTF=0.8 #FNR - Thick Forest
        self.init_dict()
    
    
    def init_dict(self,x = 0,y = 0):
        #initial probability of finding success
        unvisitedSucceeds= (1-self.fnrH)+(1-self.fnrF)+(1-self.fnrTF)
        initProb=(1/(self.n*self.n))*unvisitedSucceeds 
        for i in range(self.n):
            for j in range(self.m):
                self.dKprob[i,j]=[initProb,"NA"]
        
    def Planning(self,source,target):
        
        a=AStarResolver(self.Klst,self.n,self.m,self.Klst)
        path=a.Solve(source[0],source[1],target[0],target[1])
        if len(path)==0:
            self.updateKnowledgeBase(target[0],target[1],"NA","B")
        return path
    
    def Execution(self,path):
        compLoop=True
        bcell=(0,0)
        ubcell=(0,0)
        movement=0
        MaxprobChangedFlag=False
        for item in range(len(path)):
            if len(path)>1 and item!=0:
                x=path[item][0]
                y=path[item][1]
                if self.Flst[x][y]==-1:
                    self.Klst[x][y]=-1
                    self.updateKnowledgeBase(x,y,"NA","B")
                    compLoop=False
                    bcell=path[item-1]
                    movement=movement+1
                    break

                else:

                    
                    ubcell=(x,y)
                    if (x,y) in self.dKprob:
                        initProb=self.dKprob[x,y]
                        initProb=self.dKprob[x,y][0]
                        terrain=self.dKprob[x,y][1]
                        if terrain =="NA":#First time you are discovering the terrain
                            
                            self.dKprob[x,y]=[initProb,self.IdentifyTerrain(self.Flst[x][y])]
                            if item!=(len(path)-1):
                                self.SucceedingProbability(x,y)
                            
                        else:
                            self.dKprob[x,y]=[initProb,self.IdentifyTerrain(self.Flst[x][y])]
                    movement=movement+1
                    
                    maxed=self.TargetwithMaxProb(ubcell,False)
                    if maxed!=path[len(path)-1]:
                        MaxprobChangedFlag=True
                        compLoop=False
                        break
                        

                
        x=ubcell[0]
        y=ubcell[1]
        if compLoop==False:# cell was blocked
    
            if MaxprobChangedFlag==True:
                
                return ubcell,True,movement,MaxprobChangedFlag
            else:
                
                return bcell,True,movement,False
            
        elif ubcell==self.target:#cell was unblocked and find the target

            val=self.generateFalseTarget()
            targetFound=self.TargetMeasure(val,self.IdentifyTerrain(self.Flst[x][y]))
            if targetFound==True:
                return (0,0),False,movement,False
            else:
                self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
                self.SucceedingProbability(x,y)
                return ubcell,True,movement,False
        else:# cell was unblocked and target not found
            self.updateKnowledgeBase(x,y,self.IdentifyTerrain(self.Flst[x][y]),"UB")
            self.SucceedingProbability(x,y)
            return ubcell,True,movement,False#last element
    
    #1st argument- x
    #2nd argument- y
    #3rd argument- terrain
    #4th Condition- Blocked,Unblocked
    def updateKnowledgeBase(self,x,y,terrain,cond):
        
        initProb=self.dKprob[x,y][0]
        fnrH=0.5
        fnrF=0.2
        fnrTF=0.8
        if cond=="B":
            sumt=0
            self.dKprob[x,y]=[0,terrain]
            for i in range(self.n):
                for j in range(self.m):
                    sumt=sumt+self.dKprob[i,j][0]
            
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]
                        self.dKprob[i,j]=[probij/sumt,terraini]
            
        elif cond=="UB":
            if terrain =="F":
                probxy=initProb*fnrF
            elif terrain =="H":
                probxy=initProb*fnrH
            elif terrain=="TF":
                probxy=initProb*fnrTF
            sumt=0
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        
                        sumt=sumt+self.dKprob[i,j][0]
            probxy=probxy/(sumt+probxy)
            self.dKprob[x,y]=[probxy,terrain]
            sumt=sumt+probxy
            for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        probij=self.dKprob[i,j][0]
                        terraini=self.dKprob[i,j][1]  
                        self.dKprob[i,j]=[probij/sumt,terraini]
        
        #Agent 8
        # We are taking account of manhattan distance and updating probability of whole grid accordingly
        for i in range(self.n):
                for j in range(self.m):
                    if (i,j)!=(x,y):
                        manDist=abs(i-x)+abs(j-y)
                        prevProb=self.dKprob[i,j][0]
                        terrain=self.dKprob[i,j][1]
                        Z=manDist/(2*self.n)
                        self.dKprob[i,j]=[prevProb/Z,terrain]
        
    def SucceedingProbability(self,x,y):
        initprob=self.dKprob[x,y][0]
        terrain=self.dKprob[x,y][1]
        newProb=initprob
        if terrain=="H":
            newProb=initprob*(1-self.fnrH)
        elif terrain=="F":
            newProb=initprob*(1-self.fnrF)
        else:
            newProb=initprob*(1-self.fnrTF)
        
        self.dKprob[x,y]=[newProb,terrain]
            
    
    def TargetwithMaxProb(self,source,cond=True):
        #cond: this condition occurs when we haven't devide with manhattan distance
        maxed=max(self.dKprob, key=self.dKprob.get)
        
        #Now we are normalizing the probability of whole grid that was caculated wth the distance factor
        x=source[0]
        y=source[1]
        if cond==True:
            
            for i in range(self.n):
                    for j in range(self.m):
                        if (i,j)!=(x,y):
                            manDist=abs(i-x)+abs(j-y)
                            prevProb=self.dKprob[i,j][0]
                            terrain=self.dKprob[i,j][1]
                            Z=manDist/(2*self.n)
                            self.dKprob[i,j]=[prevProb*Z,terrain]
       # print("Final Dict Dkprob after=", self.dKprob)
        #print("Target=",maxed)
        return maxed
        
    def IdentifyTerrain(self, terrain):
        if terrain==1:
            return "F"
        elif terrain==2:
            return "H"
        elif terrain==3:
            return "TF"
        else:
            return "F"
   
    def generateFalseTarget(self):
        return random.uniform(0,1)
    
    def TargetMeasure(self,val,terrain):
        if terrain=="F":
            if val>self.fnrF:
                return True
            else:
                return False
        elif terrain=="H":
            if val>self.fnrH:
                return True
            else:
                return False
        else:
            if val>self.fnrTF:
                return True
            else:
                return False

        
```
